# *Importaci√≥n de datos*


````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
````

````python
# Seleccionar columnas necesarias
columnas_utiles = ['Producto', 'Precio', 'Calificaci√≥n', 'Costo de env√≠o']

tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

tienda_filtrada = tienda[columnas_utiles + ['Tienda']]
tienda2_filtrada = tienda2[columnas_utiles + ['Tienda']]
tienda3_filtrada = tienda3[columnas_utiles + ['Tienda']]
tienda4_filtrada = tienda4[columnas_utiles + ['Tienda']]

# Unimos los datos filtrados
df = pd.concat([tienda_filtrada, tienda2_filtrada, tienda3_filtrada, tienda4_filtrada], ignore_index=True)
````

````python
# Agrupamos por tienda y calculamos
resumen = df.groupby('Tienda').agg({
    'Precio': ['sum', 'mean'],
    'Calificaci√≥n': 'mean',
    'Costo de env√≠o': 'mean',
    'Producto': 'count'
})

# Ajustar nombres de columnas
resumen.columns = ['Ventas Totales', 'Ticket Promedio', 'Calificaci√≥n Promedio', 'Costo Envio Promedio', 'N√∫mero de Ventas']
resumen = resumen.reset_index()
````

# üìä **1. An√°lisis de Facturaci√≥n**

````python
plt.figure(figsize=(8,5))
sns.barplot(data=resumen, x='Tienda', y='Ventas Totales', palette='viridis')
plt.title('Facturaci√≥n total por Tienda')
plt.ylabel('Ventas Totales ($)')
plt.xlabel('Tienda')

# Ajuste de l√≠mites para "zoom"
min_ventas = resumen['Ventas Totales'].min()
max_ventas = resumen['Ventas Totales'].max()
plt.ylim(min_ventas * 0.9, max_ventas * 1.05)

plt.show()
````

# **6. Utilidad Estimada**

````python
# Asegurar nombres consistentes para las columnas
df.columns = [col.strip() for col in df.columns]

# Calcular resumen por tienda
resumen_utilidad = df.groupby("Tienda").agg({
    "Precio": "sum",
    "Costo de env√≠o": "mean"
}).rename(columns={
    "Precio": "Facturaci√≥n Total",
    "Costo de env√≠o": "Costo de Env√≠o Promedio"
})

# Cantidad de ventas por tienda
resumen_utilidad ["Cantidad de Ventas"] = df["Tienda"].value_counts()

# Costo total de env√≠o estimado y utilidad estimada
resumen_utilidad ["Costo Total de Env√≠o Est."] = resumen_utilidad ["Costo de Env√≠o Promedio"] * resumen_utilidad ["Cantidad de Ventas"]
resumen_utilidad ["Utilidad Estimada"] = resumen_utilidad ["Facturaci√≥n Total"] - resumen_utilidad ["Costo Total de Env√≠o Est."]

resumen_utilidad .reset_index(inplace=True)

# Configuraci√≥n est√©tica
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

# Gr√°fico de barras m√∫ltiples
resumen_utilidad.set_index("Tienda")[["Facturaci√≥n Total", "Costo Total de Env√≠o Est.", "Utilidad Estimada"]].plot(
    kind="bar", figsize=(12, 6), color=["#4C72B0", "#DD8452", "#55A868"]
)

plt.title("Comparaci√≥n de Indicadores Econ√≥micos por Tienda", fontsize=16)
plt.ylabel("Valor en COP")
plt.xlabel("Tienda")
plt.xticks(rotation=0)
plt.legend(title="Indicadores", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

--
Al analizar la utilidad estimada (facturaci√≥n total menos costo de env√≠o), se observa que Tienda 1 genera la mayor utilidad, lo que indica un excelente balance entre ingresos por ventas y gastos log√≠sticos. Tienda 2 y Tienda 3 mantienen niveles de utilidad similares y competitivos, lo que sugiere una operaci√≥n eficiente. Por el contrario, Tienda 4, aunque presenta el costo de env√≠o m√°s bajo, tambi√©n registra la menor utilidad, lo que puede deberse a un volumen de ventas m√°s bajo o precios menos competitivos.

üìå Conclusi√≥n: Tienda 1 es la opci√≥n m√°s rentable. Tienda 4 requiere evaluaci√≥n estrat√©gica para mejorar su desempe√±o financiero.
